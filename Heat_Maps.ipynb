## `Main Heat Map code`

```python
def build_state_heatmap(state: str, branch_data_dict: dict, *,
        df_path: str = "/content/drive/MyDrive/Heat_Maps/lap_match.csv", pincode_less_path: str = "/content/drive/MyDrive/Heat_Maps/pincode_less.csv",
        geojson_dir: str = "/content/drive/MyDrive/Heat_Maps", base_output_folder: str = "/content/drive/MyDrive/Heat_Maps/Heat_Maps_output",
        exclude_pincodes: list[int] | None = None, radius_km: int = 50 ) -> None:

    if exclude_pincodes is None:
        exclude_pincodes = []

    # 1) ----------------- flat DataFrame of every branch -------------
    branches = []
    for st, cities in branch_data_dict.items():
        for ct, meta in cities.items():
            branches.append(
                {"state": st, "city": ct,
                 "latitude": meta["coordinates"][0],
                 "longitude": meta["coordinates"][1]}
            )
    branch_df = pd.DataFrame(branches)
    state_branches = branch_df[branch_df["state"] == state]
    if state_branches.empty:
        raise ValueError(f"No branches found for state '{state}'")

    # 2) ----------------- load pincode + KPI tables ------------------
    df = pd.read_csv(df_path)
    df["state_y"] = df["state_y"].str.lower()
    df = df[df["state_y"] == state.lower()]

    pincode_less_df = pd.read_csv(pincode_less_path)
    pincode_less_df["state"] = pincode_less_df["state"].str.lower()
    pincode_less_df = pincode_less_df[pincode_less_df["state"] == state.lower()]

    pincode_df = df[["pincode", "latitude", "longitude"]].copy()
    pincode_df["rad_lat"] = np.deg2rad(pincode_df["latitude"])
    pincode_df["rad_lon"] = np.deg2rad(pincode_df["longitude"])

    # 3) ----------------- build BallTree for pincode search ----------
    pin_tree = BallTree(pincode_df[["rad_lat", "rad_lon"]].values, metric="haversine")

    # 3a) find *any* pincode within radius_km from *any* branch
    rad = radius_km / 6371.0
    br_rad = np.deg2rad(state_branches[["latitude", "longitude"]].values)
    idxs = pin_tree.query_radius(br_rad, r=rad)
    pins_within = np.unique(np.concatenate(idxs))
    pincode_df = pincode_df.iloc[pins_within].copy()
    pincode_df = pincode_df[~pincode_df["pincode"].isin(exclude_pincodes)]

    # 4) ----------------- nearest‑branch & nearby‑branch columns -----
    # build BallTree on *branches*:
    branch_coords_rad = np.deg2rad(state_branches[["latitude", "longitude"]].values)
    branch_tree = BallTree(branch_coords_rad, metric="haversine")

    pin_coords_rad = pincode_df[["rad_lat", "rad_lon"]].values
    d_rad, idx = branch_tree.query(pin_coords_rad, k=len(state_branches))
    d_km = d_rad * 6371

    nearest_idx = idx[:, 0]
    pincode_df["nearest_branch"] = state_branches.iloc[nearest_idx]["city"].values
    pincode_df["distance_to_nearest_km"] = d_km[:, 0].round(2)

    within_mask = d_km <= radius_km
    pincode_df["nearby_branches"] = [
        ", ".join(state_branches["city"][within_mask[i]].tolist())
        for i in range(len(pincode_df))
    ]

    # 5) ----------------- merge KPI / scoring ------------------------
    heat_df = pd.merge(
        pincode_df,
        pincode_less_df,
        left_on="pincode", right_on="std_pin_code",
        how="left"
    )
    heat_df["score"] = heat_df["score"].fillna(0)
    heat_df = heat_df.dropna(subset=["average_growth_rate", "quanta_in_crore"])

    # 6) ----------------- build base map -----------------------------
    geojson_file = os.path.join(
        geojson_dir,
        f"{state.lower().replace(' ', '_')}_state.geojson"
    )
    boundary = json.load(open(geojson_file))
    gdf = gpd.GeoDataFrame.from_features(boundary["features"])
    if gdf.crs is None:
        gdf = gdf.set_crs(epsg=4326)
    cent = gdf.geometry.union_all().centroid
    fmap = folium.Map(location=[cent.y, cent.x], zoom_start=7)

    HeatMap(heat_df[["latitude", "longitude", "score"]].values.tolist()
            ).add_to(fmap)

    # pin markers
    for _, r in heat_df.iterrows():
        folium.CircleMarker(
            [r["latitude"], r["longitude"]],
            radius=4, color="blue", fill=True, fill_opacity=0.5,
            popup=folium.Popup(
                f"<b>Pincode:</b> {r['pincode']}<br>"
                f"<b>Nearby Branches:</b> {r['nearby_branches']}<br>"
                f"<b>Nearest Branch:</b> {r['nearest_branch']}<br>"
                f"<b>Distance:</b> {r['distance_to_nearest_km']} km<br>"
                f"<b>Avg Growth:</b> {round(r['average_growth_rate']*100,2)} %<br>"
                f"<b>Quanta (Cr):</b> {r['quanta_in_crore']}",
                max_width=300
            )
        ).add_to(fmap)

    # branch markers & circles
    for _, br in state_branches.iterrows():
        folium.Marker(
            [br["latitude"], br["longitude"]],
            popup=f"Branch in {br['city']}"
        ).add_to(fmap)
        folium.Circle(
            [br["latitude"], br["longitude"]],
            radius=radius_km*1000, color="blue", fill=False, fill_opacity=0.07
        ).add_to(fmap)

    folium.GeoJson(
        boundary,
        style_function=lambda _: {"color": "black", "weight": 2, "fill": False},
        name=state
    ).add_to(fmap)

    folium.LayerControl(collapsed=True).add_to(fmap)

    # 7) ----------------- save outputs --------------------------------
    out_dir='/content/drive/MyDrive/Heat_Maps/Heat_Maps_output'
    html_path = os.path.join(out_dir, f"heatmap_{state.lower().replace(' ', '_')}.html")
    csv_path  = os.path.join(out_dir, f"{state.lower().replace(' ', '_')}_detail.csv")

    fmap.save(html_path)
    heat_df[[
        "pincode",
        "nearby_branches",
        "distance_to_nearest_km",
        "average_growth_rate",
        "quanta_in_crore"
    ]].to_csv(csv_path, index=False)

    print(f"✅ HTML: {html_path}")
    print(f"✅  CSV: {csv_path}")
```

## Branch Wise Folium Heat Maps

```python
def build_branch_heatmap(state, branch_city, branch_data_dict,
        *, df_path: str = "/content/drive/MyDrive/Heat_Maps/lap_match.csv", pincode_less_path: str = "/content/drive/MyDrive/Heat_Maps/pincode_less.csv",
        geojson_dir: str = "/content/drive/MyDrive/Heat_Maps", base_output_folder: str = "/content/drive/MyDrive/Heat_Maps/Heat_Maps_output",
        exclude_pincodes: list[int]      = None, radius_km: int= 50 ) -> None:

    if exclude_pincodes is None:
        exclude_pincodes = []
    # ----------------------- 1.  build tiny branch DataFrame ---------
    rows = []
    for st, cities in branch_data_dict.items():
        for ct, meta in cities.items():
            rows.append(
                {"state": st, "city": ct,
                 "latitude": meta["coordinates"][0],
                 "longitude": meta["coordinates"][1]}
            )
    branch_df = pd.DataFrame(rows)

    state_df   = branch_df[branch_df["state"] == state]
    branch_row = state_df[state_df["city"] == branch_city]
    if branch_row.empty:
        raise ValueError(f"‘{branch_city}’ not found in {state}")
    branch_row = branch_row.iloc[0]
    focus_branch_df = branch_row.to_frame().T

    # ----------------------- 2.  load pincode & KPI CSVs -------------
    df = pd.read_csv(df_path)
    df["state_y"] = df["state_y"].str.lower()
    df = df[df["state_y"] == state.lower()]

    pincode_less_df = pd.read_csv(pincode_less_path)
    pincode_less_df["state"] = pincode_less_df["state"].str.lower()
    pincode_less_df = pincode_less_df[pincode_less_df["state"] == state.lower()]

    pincode_data = df[["pincode", "latitude", "longitude"]].copy()
    pincode_data["rad_lat"] = np.deg2rad(pincode_data["latitude"])
    pincode_data["rad_lon"] = np.deg2rad(pincode_data["longitude"])
    pin_tree = BallTree(pincode_data[["rad_lat", "rad_lon"]].values, metric="haversine")

    # ----------------------- 3.  find pins within radius ------------
    rad = radius_km / 6371.0
    br_lat_rad = np.deg2rad(branch_row["latitude"])
    br_lon_rad = np.deg2rad(branch_row["longitude"])
    idx = pin_tree.query_radius([[br_lat_rad, br_lon_rad]], r=rad)[0]
    pincode_data_filtered = pincode_data.iloc[idx].copy()

    pincode_data_filtered = pincode_data_filtered[~pincode_data_filtered["pincode"].isin(exclude_pincodes)]

    # ----------------------- 4.  annotate pins -----------------------
    pincode_data_filtered["nearest_branch"] = branch_city
    d_rad, _ = BallTree([[br_lat_rad, br_lon_rad]], metric="haversine").query(
                        pincode_data_filtered[["rad_lat", "rad_lon"]], k=1)
    pincode_data_filtered["distance_to_nearest_km"] = (d_rad.flatten() * 6371).round(2)
    pincode_data_filtered["nearby_branches"] = np.where(
        pincode_data_filtered["distance_to_nearest_km"] <= radius_km,
        branch_city, ""
    )

    # ----------------------- 5.  merge KPI data ---------------------
    heatmap_df = pd.merge(
        pincode_data_filtered,
        pincode_less_df,
        left_on="pincode", right_on="std_pin_code",
        how="left"
    )
    heatmap_df["score"] = heatmap_df["score"].fillna(0)
    heatmap_df = heatmap_df.dropna(subset=["average_growth_rate", "quanta_in_crore"])

    # ----------------------- 6.  build map ---------------------------
    geojson_file = os.path.join(
        geojson_dir,
        f"{state.lower().replace(' ', '_')}_state.geojson"
    )
    state_boundary = json.load(open(geojson_file))
    gdf = gpd.GeoDataFrame.from_features(state_boundary["features"])
    if gdf.crs is None:
        gdf = gdf.set_crs(epsg=4326)
    centroid = gdf.geometry.union_all().centroid

    fol_map = folium.Map(location=[centroid.y, centroid.x], zoom_start=7)

    HeatMap(heatmap_df[["latitude", "longitude", "score"]].values.tolist()
            ).add_to(fol_map)

    for _, row in heatmap_df.iterrows():
        popup = folium.Popup(
            f"<b>Pincode:</b> {row['pincode']}<br>"
            f"<b>Nearest Branch:</b> {row['nearest_branch']}<br>"
            f"<b>Distance:</b> {row['distance_to_nearest_km']} km<br>"
            f"<b>Average Growth:</b> {round(row['average_growth_rate']*100,2)} %<br>"
            f"<b>Quanta (Cr):</b> {row['quanta_in_crore']}",
            max_width=300
        )
        folium.CircleMarker(
            [row["latitude"], row["longitude"]],
            radius=5, color="blue", fill=True, fill_opacity=0.5,
            popup=popup
        ).add_to(fol_map)

    folium.Marker(
        [branch_row["latitude"], branch_row["longitude"]],
        popup=f"Branch in {branch_city}"
    ).add_to(fol_map)

    folium.Circle(
        [branch_row["latitude"], branch_row["longitude"]],
        radius=radius_km*1000, color="blue", fill=False, fill_opacity=0.1
    ).add_to(fol_map)

    folium.GeoJson(
        state_boundary,
        style_function=lambda _: {"color": "black", "weight": 2, "fill": False},
        name=state
    ).add_to(fol_map)

    folium.LayerControl(collapsed=True).add_to(fol_map)

    # ----------------------- 7.  save outputs ------------------------
    out_dir = os.path.join(
        base_output_folder,
        "Branch_Wise",
        state.lower().replace(" ", "_"),
        branch_city.replace(" ", "_")
    )
    os.makedirs(out_dir, exist_ok=True)

    html_path = os.path.join(out_dir, f"heatmap_{branch_city.replace(' ', '_')}.html")
    csv_path  = os.path.join(out_dir, f"{branch_city.replace(' ', '_')}_detail.csv")

    fol_map.save(html_path)
    heatmap_df[[
        "pincode",
        "nearby_branches",
        "distance_to_nearest_km",
        "average_growth_rate",
        "quanta_in_crore"
    ]].to_csv(csv_path, index=False)

    print(f"✅ HTML: {html_path}")
    print(f"✅  CSV: {csv_path}")
```

```python
import pandas as pd
import numpy as np
import geopandas as gpd
import folium
from sklearn.neighbors import BallTree
import json
import os

def build_state_heatmap(state: str, branch_data_dict: dict, *,
        df_path: str = "/content/drive/MyDrive/Heat_Maps/lap_match.csv",
        pincode_less_path: str = "/content/drive/MyDrive/Heat_Maps/pincode_less.csv",
        geojson_dir: str = "/content/drive/MyDrive/Heat_Maps",
        base_output_folder: str = "/content/drive/MyDrive/Heat_Maps/Heat_Maps_output",
        exclude_pincodes: list[int] | None = None,
        radius_km: int = 50) -> None:

    if exclude_pincodes is None:
        exclude_pincodes = []

    # List to collect heat_df for all states
    all_heat_dfs = []

    # 1) ----------------- flat DataFrame of every branch -------------
    branches = []
    for st, cities in branch_data_dict.items():
        for ct, meta in cities.items():
            branches.append(
                {"state": st, "city": ct,
                 "latitude": meta["coordinates"][0],
                 "longitude": meta["coordinates"][1]}
            )
    branch_df = pd.DataFrame(branches)
    state_branches = branch_df[branch_df["state"] == state]
    if state_branches.empty:
        raise ValueError(f"No branches found for state '{state}'")

    # 2) ----------------- load pincode + KPI tables ------------------
    df = pd.read_csv(df_path)
    df["state_y"] = df["state_y"].str.lower()
    df = df[df["state_y"] == state.lower()]

    pincode_less_df = pd.read_csv(pincode_less_path)
    pincode_less_df["state"] = pincode_less_df["state"].str.lower()
    pincode_less_df = pincode_less_df[pincode_less_df["state"] == state.lower()]

    pincode_df = df[["pincode", "latitude", "longitude"]].copy()
    pincode_df["rad_lat"] = np.deg2rad(pincode_df["latitude"])
    pincode_df["rad_lon"] = np.deg2rad(pincode_df["longitude"])

    # 3) ----------------- build BallTree for pincode search ----------
    pin_tree = BallTree(pincode_df[["rad_lat", "rad_lon"]].values, metric="haversine")

    # 3a) find *any* pincode within radius_km from *any* branch
    rad = radius_km / 6371.0
    br_rad = np.deg2rad(state_branches[["latitude", "longitude"]].values)
    idxs = pin_tree.query_radius(br_rad, r=rad)
    pins_within = np.unique(np.concatenate(idxs))
    pincode_df = pincode_df.iloc[pins_within].copy()
    pincode_df = pincode_df[~pincode_df["pincode"].isin(exclude_pincodes)]

    # 4) ----------------- nearest-branch & nearby-branch columns -----
    branch_coords_rad = np.deg2rad(state_branches[["latitude", "longitude"]].values)
    branch_tree = BallTree(branch_coords_rad, metric="haversine")

    pin_coords_rad = pincode_df[["rad_lat", "rad_lon"]].values
    d_rad, idx = branch_tree.query(pin_coords_rad, k=len(state_branches))
    d_km = d_rad * 6371

    nearest_idx = idx[:, 0]
    pincode_df["nearest_branch"] = state_branches.iloc[nearest_idx]["city"].values
    pincode_df["distance_to_nearest_km"] = d_km[:, 0].round(2)

    within_mask = d_km <= radius_km
    pincode_df["nearby_branches"] = [
        ", ".join(state_branches["city"][within_mask[i]].tolist())
        for i in range(len(pincode_df))
    ]

    # 5) ----------------- merge KPI / scoring ------------------------
    heat_df = pd.merge(
        pincode_df,
        pincode_less_df,
        left_on="pincode", right_on="std_pin_code",
        how="left"
    )
    heat_df["score"] = heat_df["score"].fillna(0)
    heat_df = heat_df.dropna(subset=["average_growth_rate", "quanta_in_crore"])
    heat_df["state"] = state  # Add state column to identify source state

    # Append to list for combined CSV
    all_heat_dfs.append(heat_df)

    # 6) ----------------- build base map -----------------------------
    geojson_file = os.path.join(
        geojson_dir,
        f"{state.lower().replace(' ', '_')}_state.geojson"
    )
    boundary = json.load(open(geojson_file))
    gdf = gpd.GeoDataFrame.from_features(boundary["features"])
    if gdf.crs is None:
        gdf = gdf.set_crs(epsg=4326)
    cent = gdf.geometry.union_all().centroid
    fmap = folium.Map(location=[cent.y, cent.x], zoom_start=7)

    HeatMap(heat_df[["latitude", "longitude", "score"]].values.tolist()
            ).add_to(fmap)

    # pin markers
    for _, r in heat_df.iterrows():
        folium.CircleMarker(
            [r["latitude"], r["longitude"]],
            radius=4, color="blue", fill=True, fill_opacity=0.5,
            popup=folium.Popup(
                f"<b>Pincode:</b> {r['pincode']}<br>"
                f"<b>Nearby Branches:</b> {r['nearby_branches']}<br>"
                f"<b>Nearest Branch:</b> {r['nearest_branch']}<br>"
                f"<b>Distance:</b> {r['distance_to_nearest_km']} km<br>"
                f"<b>Avg Growth:</b> {round(r['average_growth_rate']*100,2)} %<br>"
                f"<b>Quanta (Cr):</b> {r['quanta_in_crore']}",
                max_width=300
            )
        ).add_to(fmap)

    # branch markers & circles
    for _, br in state_branches.iterrows():
        folium.Marker(
            [br["latitude"], br["longitude"]],
            popup=f"Branch in {br['city']}"
        ).add_to(fmap)
        folium.Circle(
            [br["latitude"], br["longitude"]],
            radius=radius_km*1000, color="blue", fill=False, fill_opacity=0.07
        ).add_to(fmap)

    folium.GeoJson(
        boundary,
        style_function=lambda _: {"color": "black", "weight": 2, "fill": False},
        name=state
    ).add_to(fmap)

    folium.LayerControl(collapsed=True).add_to(fmap)

    # 7) ----------------- save outputs --------------------------------
    out_dir = base_output_folder
    html_path = os.path.join(out_dir, f"heatmap_{state.lower().replace(' ', '_')}.html")
    csv_path = os.path.join(out_dir, f"{state.lower().replace(' ', '_')}_detail.csv")

    fmap.save(html_path)
    heat_df[[
        "pincode",
        "nearby_branches",
        "distance_to_nearest_km",
        "average_growth_rate",
        "quanta_in_crore",
        "state"  # Include state in per-state CSV
    ]].to_csv(csv_path, index=False)

    # Save combined CSV with all pincodes
    if all_heat_dfs:
        combined_df = pd.concat(all_heat_dfs, ignore_index=True)
        combined_csv_path = os.path.join(out_dir, "all_pincodes.csv")
        combined_df[[
            "pincode",
            "nearby_branches",
            "distance_to_nearest_km",
            "average_growth_rate",
            "quanta_in_crore",
            "state"
        ]].to_csv(combined_csv_path, index=False)
        print(f"✅ Combined CSV: {combined_csv_path}")

    print(f"✅ HTML: {html_path}")
    print(f"✅ CSV: {csv_path}")
```
